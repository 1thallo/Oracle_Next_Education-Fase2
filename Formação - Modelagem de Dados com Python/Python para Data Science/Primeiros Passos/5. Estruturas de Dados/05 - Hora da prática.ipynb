#### Quest√£o 5
Fa√ßa um programa que, ao inserir um n√∫mero qualquer, criar√° uma lista contendo todos os n√∫meros primos entre 1 e o n√∫mero digitado.

````python
lista_primos = []

numeroQ5 = int(input("Insira o n√∫mero: "))

for num in range(2, numeroQ5): 
        primo = True
        
        for divisor in range(2, num): 
            if num % divisor == 0:
                primo = False
                break
            
        if primo == True:
            lista_primos.append(num)

print(f"N√∫meros primos entre 1 e {numeroQ5}: {lista_primos}")
````

#### Quest√£o 7
Para um estudo envolvendo o n√≠vel de multiplica√ß√£o de bact√©rias em uma col√¥nia, foi coletado o n√∫mero de bact√©rias multiplicadas por dia, com base em um processo de duplica√ß√£o di√°ria e pode ser observado a seguir: `[1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]`. Tendo esses valores, fa√ßa um c√≥digo que gere uma lista contendo o percentual de crescimento de bact√©rias por dia, comparando o n√∫mero de bact√©rias em cada dia com o n√∫mero de bact√©rias do dia anterior. Dica: para calcular o percentual de crescimento usamos a seguinte equa√ß√£o: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
# listaa
bacterias = [1.2, 1.1, 2.1, 2.0, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]

percentual_crescimento = []

for i in range(1, len(bacterias)):
    valor_anterior = bacterias[i - 1]
    valor_atual = bacterias[i]
    crescimento = (valor_atual - valor_anterior) / valor_anterior * 100
    percentual_crescimento.append(crescimento)

# 2 casas decimais
percentual_arredondado = []
for valor in percentual_crescimento:
    valor_formatado = round(valor, 2)
    percentual_arredondado.append(valor_formatado)

# formatar % em cada valor
percentual_com_sinal = []
for valor in percentual_arredondado:
    if valor >= 0:
        percentual_formatado = "+" + str(valor) + "%"
        percentual_com_sinal.append(percentual_formatado)
    else:
        percentual_formatado = str(valor) + "%"
        percentual_com_sinal.append(percentual_formatado)

## 
print("Percentual de crescimento por dia:", percentual_com_sinal)
````

#### Quest√£o 8
Para uma sele√ß√£o de produtos aliment√≠cios, precisamos separar o conjunto de IDs dados por n√∫meros inteiros, sabendo que os produtos com ID par s√£o doces e os com ID √≠mpar s√£o amargos. Monte um c√≥digo que colete 10 IDs. Depois, calcule e mostre a quantidade de produtos doces e amargos.

````python
lista_doces = []
lista_amargos = []

for i in range(1,11):
    while True:
        try:
            id_produto = int(input(f"Escreva o ID do produto {i}: "))
            
            if id_produto <= 0:
                print("Digite um n√∫mero positivo.")
                continue
            
            if id_produto in lista_doces or id_produto in lista_amargos:
                print("ID j√° inserido!")
                continue
    
            if id_produto % 2 == 0:
                lista_doces.append(id_produto)
            else:
                lista_amargos.append(id_produto)
            break
        except ValueError:
            print("Digite um n√∫mero, por favor.")

print(f"""\n------- Relat√≥rio Geral -------\n
Produtos doces: {lista_doces}
Quantidade: {len(lista_doces)}

Produtos Amargos: {lista_amargos}
Quantidade: {len(lista_amargos)}

Quantidade Total: {len(lista_amargos) + len(lista_doces)}
""")
````

#### Quest√£o 9
Desenvolva um programa que informa a nota de um aluno de acordo com suas respostas. Ele deve pedir a resposta de um aluno para cada quest√£o e √© preciso verificar se a resposta foi igual ao gabarito. Cada quest√£o vale um ponto e existem alternativas:  A, B, C ou D.
```
Gabarito da Prova:
01 - D
02 - A
03 - C
04 - B
05 - A
06 - D
07 - C
08 - C
09 - A
10 - B
```

````python
gabarito = {
    1: 'D', 2: 'A', 3: 'C', 4: 'B', 5: 'A',
    6: 'D', 7: 'C', 8: 'C', 9: 'A', 10: 'B'
}

respostas = {}
nota = 0

for i in range(1, len(gabarito) + 1):
    while True:
        try:
            resposta = input(f"Resposta da Quest√£o {i}: ").upper().strip()
            
            if not resposta.isalpha():
                print("Erro: Digite uma LETRA da alternativa!")
                continue

            if len(resposta) != 1:
                print("Digite s√≥ um caractere por resposta!")
                continue

            if resposta not in ('A', 'B', 'C', 'D'):
                print("Insira uma alternativa existente - (A, B, C ou D).")
                continue

            respostas[i] = resposta  # salva a resposta no dicion√°rio

            if resposta == gabarito[i]:
                nota += 1

            break  # sai do while qnd a resposta √© v√°lida
        except Exception:
            print("Erro inesperado! Tente novamente.")

print("""
-------------------------------------------------------
Gabarito da prova:       {}
Respostas do aluno:      {}
Nota final:              {} de {}
-------------------------------------------------------
""".format(gabarito, respostas, nota, len(gabarito)))
````

#### Quest√£o 10
Um instituto de meteorologia deseja fazer um estudo de temperatura m√©dia de cada m√™s do ano. Para isso, voc√™ precisa fazer um c√≥digo que colete e armazene essas temperaturas m√©dias em uma lista. Depois, calcule a m√©dia anual das temperaturas e mostre todas as temperaturas acima da m√©dia anual, e em que m√™s elas ocorreram, mostrando os meses por extenso: Janeiro, Fevereiro, etc.

````python
meses = [
    "Janeiro", "Fevereiro", "Mar√ßo", "Abril", "Maio", "Junho",
    "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"
]

temperaturas = []

for i in range(12):
    while True:
        try:
            temperatura = float(input(f"Temperatura do m√™s {meses[i]}: "))
            temperaturas.append(temperatura)
            break
        except ValueError:
            print("Erro: Insira um n√∫mero v√°lido!")

#media
media_anual = sum(temperaturas) / len(temperaturas)

#indices para pegar valores max e min da lista
indice_max = temperaturas.index(max(temperaturas))
indice_min = temperaturas.index(min(temperaturas))

##
print(f"""
---------- Relat√≥rio ----------
M√©dia anual: {media_anual:.2f} ¬∫C
M√™s mais quente: {meses[indice_max]} ({temperaturas[indice_max]:.2f} ¬∫C)
M√™s mais frio: {meses[indice_min]} ({temperaturas[indice_min]:.2f} ¬∫C)
""")

print("------- Temperaturas acima da m√©dia anual -------")
for i in range(12):
    if temperaturas[i] > media_anual:
        print(f"{meses[i]} - {temperaturas[i]:.2f} ¬∫C")
````

#### Quest√£o 12
Uma pesquisa de mercado foi feita para decidir qual design de uma marca infantil mais agrada crian√ßas. A pesquisa foi feita e o votos computados podem ser observados abaixo:
```
'''
Tabela de votos da marca
Design 1 - 1334 votos
Design 2 - 982 votos
Design 3 - 1751 votos
Design 4 - 210 votos
Design 5 - 1811 votos
'''
```
Adapte os dados fornecidos a voc√™ para uma estrutura de dicion√°rio e a partir dele, informe o design vencedor e a porcentagem de votos recebidos.


````python
votos = {'D1': 1334, 'D2': 982, 'D3': 1751, 'D4': 210, 'D5': 1811}

vencedor = max(votos, key=votos.get)
total_votos = sum(votos.values())

print(f""" ------------- Relat√≥rio de Votos -------------
Total de votos: {total_votos} votos
Design Vencedor: {vencedor} ({votos[vencedor]} votos)
\n------------ Porcentagem de votos por design ------------ """)

for design, qtd in votos.items():
    percentual = (qtd / total_votos) * 100
    print(f"{design}: {percentual:.2f}%")

````

Os funcion√°rios de um setor da empresa que voc√™ trabalha v√£o receber um abono correspondente a 10% do seu sal√°rio devido ao √≥timo desempenho do time. O setor financeiro solicitou sua ajuda para a verifica√ß√£o das consequ√™ncias financeiras que esse abono ir√° gerar nos recursos. Assim, foi encaminhada para voc√™ uma lista com os sal√°rios que receber√£o o abono: `[1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]`. O abono de cada funcion√°rio n√£o pode ser inferior a 200. Em c√≥digo, transforme cada um dos sal√°rios em chaves de um dicion√°rio e o abono de cada sal√°rio no elemento. Depois, informe o total de gastos com o abono, quantos funcion√°rios receberam o abono m√≠nimo e qual o maior valor de abono fornecido.


````python
salarios_abono = [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]

dict_abono = {}
total_abono = 0
min_abono = 0
max_abono = 0

for salario in salarios_abono:
    abono = salario * 0.1   #10% do salario
    
    if abono < 200:
        abono = 200
        
    dict_abono[salario] = abono
    
for abono in dict_abono.values():
    if abono == 200:
        min_abono += 1
    
    if abono > max_abono:
        max_abono = abono
    total_abono += abono

## perfumaria kkkk ü•∏‚òùÔ∏è
print(f""" ----------- Relat√≥rio de Abonos -----------
   Sal√°rio  |   Abono
-------------------------- """)
for salario, abono in dict_abono.items():
    print(f"R$ {salario:7.2f}  |  R$ {abono:.2f}")

print(f"""\n----------- Resumo Geral -----------
Total de gasto: R$ {total_abono:.2f}
Funcion√°rios com abono m√≠nimo: {min_abono} funcion√°rios
Maior abono: {max_abono:.2f}
""")
    
    
````

#### Quest√£o 14
Uma equipe de cientistas de dados est√° estudando a diversidade biol√≥gica em uma floresta. A equipe fez a coleta de informa√ß√µes sobre o n√∫mero de esp√©cies de plantas e animais em cada √°rea da floresta e armazenaram essas informa√ß√µes em um dicion√°rio. Nele, a chave descreve a √°rea dos dados e os valores nas listas correspondem √†s esp√©cies de plantas e animais nas √°reas, respectivamente.
```
{'√Årea Norte': [2819, 7236],
 '√Årea Leste': [1440, 9492],
 '√Årea Sul': [5969, 7496],
 '√Årea Oeste': [14446, 49688],
 '√Årea Centro': [22558, 45148]}
```
Escreva um c√≥digo para calcular a m√©dia de esp√©cies por √°rea e identificar a √°rea com a maior diversidade biol√≥gica. Dica: use as fun√ß√µes built-in [`sum()`](https://docs.python.org/3/library/functions.html#sum) e [`len()`](https://docs.python.org/3/library/functions.html#len).


````python
coleta_infos = {
    '√Årea Norte': [2819, 7236],
    '√Årea Leste': [1440, 9492],
    '√Årea Sul': [5969, 7496],
    '√Årea Oeste': [14446, 49688],
    '√Årea Centro': [22558, 45148]
}

medias_por_area = {}

for area, especies in coleta_infos.items():
    media = sum(especies) / len(especies)
    medias_por_area[area] = media

area_mais_diversa = max(medias_por_area, key=medias_por_area.get)

##
print("----- M√©dia de Esp√©cies por √Årea -----")
for area, media in medias_por_area.items():
    print(f"{area}: {media:.2f} esp√©cies em m√©dia")

print(f"\n√Årea com maior diversidade biol√≥gica: {area_mais_diversa} ({medias_por_area[area_mais_diversa]:.2f} esp√©cies em m√©dia)")
````

#### Quest√£o 15
O setor de RH da sua empresa te pediu uma ajuda para analisar as idades dos funcion√°rios de 4 setores da empresa. Para isso, ele te forneceu os seguintes dados:
```
{'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}
``` 
Sabendo que cada setor tem 10 funcion√°rios, construa um c√≥digo que calcule a m√©dia de idade de cada setor, a idade m√©dia geral entre todos os setores e quantas pessoas est√£o acima da idade m√©dia geral.


````python
idades_por_setor = {'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}

pessoas_acima_da_media = 0
total_idades = 0

print("----- M√©dia de idades por setor -----")
for key_setor, val_idades in idades_por_setor.items():
    media_idades = sum(val_idades) / len(val_idades)

    print(f"{key_setor}: {media_idades}")
    
    total_idades += sum(val_idades)

media_total = total_idades / (len(val_idades) * len(idades_por_setor))
print(f"\nM√©dia geral: {media_total}")

for setor, idade in idades_por_setor.items():
    for _ in idade:
        if _ > media_total:
            pessoas_acima_da_media += 1

print(f"\n{pessoas_acima_da_media} pessoas est√£o acima da m√©dia geral de idade!")
````
