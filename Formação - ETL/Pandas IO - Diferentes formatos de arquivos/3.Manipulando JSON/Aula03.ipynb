Leitura Arquivo JSON

```python
import pandas as pd
import os

caminho = os.path.join('..', 'data', 'pacientes.json')

dados = pd.read_json(caminho)

dados

```

```python
caminho_2 = os.path.join('..', 'data', 'pacientes_2.json')

dados_2 = pd.read_json(caminho_2)

dados_2
```

### Normalizando JSON

```python
df_normalizado = pd.json_normalize(dados_2['Pacientes'])
df_normalizado
```

### Escrevendo Arquivos JSON

```python
df_normalizado.to_json('historico_pacientes_normalizados.json')
```
